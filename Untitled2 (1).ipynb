# **Tratamento dos Dados: Datas, Nulos, Duplicatas e Outliers**

```python
# Converter colunas de data para o tipo datetime
df['Order.Date'] = pd.to_datetime(df['Order.Date'], errors='coerce')
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'], errors='coerce')

df.info()
```

```python
# Verificar a quantidade de valores ausentes em cada coluna
df.isnull().sum().sort_values(ascending=False)
```

```python
# Remover linhas com valores ausentes
df.dropna(inplace=True)
```

```python
# Confirmar que todos os dados ausentes foram removidos
df.isnull().sum()
```

```python
# Verificar quantas linhas duplicadas existem
df.duplicated().sum()
```

```python
# Verificar colunas categóricas
colunas_categoricas = df.select_dtypes(include='object').columns
print(colunas_categoricas)
```

```python
# Verifica valores únicos de cada coluna categórica
for coluna in colunas_categoricas:
    print(f"Coluna: {coluna}")
    print(df[coluna].unique())
    print("------")
```

```python
# Padronizar texto das colunas categóricas (strip e lower)
for coluna in colunas_categoricas:
    df[coluna] = df[coluna].str.strip().str.lower()
```

```python
# Capitalizar mantendo as iniciais maiúsculas
for coluna in colunas_categoricas:
    df[coluna] = df[coluna].str.strip().str.title()
```

```python
# Visualizar outliers na coluna 'Profit' usando boxplot
sns.boxplot(x=df['Profit'])
plt.title('Boxplot do Lucro')
plt.show()
```

```python
# Remover valores de 'Profit' considerados outliers (fora do intervalo -4500 a 4500)
df = df[df['Profit'].between(-4500, 4500)]
```

```python
# Visualizar a distribuição da coluna 'Profit' após remoção dos outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Profit'])
plt.title('Boxplot do Lucro (após remoção de outliers)')
plt.show()
```

```python
# Extrair o ano da data do pedido e criar uma nova coluna 'Ano'
df['Ano'] = df['Order.Date'].dt.year

# Calcular o tempo de entrega em dias (diferença entre data de envio e data do pedido)
df['Lead_Time'] = (df['Ship.Date'] - df['Order.Date']).dt.days

# Mostrar as primeiras linhas para conferir as novas colunas
df.head()
```

```python
# Renomear colunas para nomes em português e mais claros, facilitando a leitura e a análise
df.rename(columns={
    'Category': 'Categoria',
    'City': 'Cidade',
    'Country': 'País',
    'Customer.ID': 'ID_Cliente',
    'Customer.Name': 'Nome_Cliente',
    'Discount': 'Desconto',
    'Market': 'Mercado',
    'Order.Date': 'Data_Pedido',
    'Order.ID': 'ID_Pedido',
    'Order.Priority': 'Prioridade_Pedido',
    'Product.ID': 'ID_Produto',
    'Product.Name': 'Nome_Produto',
    'Profit': 'Lucro',
    'Quantity': 'Quantidade',
    'Region': 'Região',
    'Row.ID': 'ID_Linha',
    'Sales': 'Vendas',
    'Segment': 'Segmento',
    'Ship.Date': 'Data_Envio',
    'Ship.Mode': 'Modo_Envio',
    'Shipping.Cost': 'Custo_Envio',
    'State': 'Estado',
    'Sub.Category': 'Subcategoria',
    'Year': 'Ano',
    'Market2': 'Mercado2',
    'weeknum': 'Semana_Ano',
    '记录数': 'Qtd_Registro', # Parece um campo em chinês, mantive tradução genérica
    'Lead_Time': 'Tempo_Entrega_Dias'
}, inplace=True)

print(df.columns)
```

#  **Análise Exploratória (EDA)**

```python
# Agrupar os dados por segmento e calcular métricas de lucro
lucro_segmento = df.groupby('Segmento').agg({
    'Lucro': ['sum', 'mean', 'count']
}).reset_index()

# Renomear as colunas para nomes mais descritivos
lucro_segmento.columns = ['Segmento', 'Lucro_Total', 'Lucro_Medio', 'Qtd_Pedidos']

# Exibir o resultado ordenado pelo lucro total (do maior para o menor)
lucro_segmento.sort_values(by='Lucro_Total', ascending=False)
```

```python
# Dados
segmentos = ['Consumer', 'Corporate', 'Home Office']
lucro_total = [734502.94816, 432808.35266, 277009.18056]
lucro_medio = [27.703502, 28.053432, 29.648847]

# Criando a figura e os eixos
fig, ax1 = plt.subplots()

# Cor e barra do Lucro Total (eixo primário)
ax1.bar(segmentos, lucro_total, color='skyblue', label='Lucro Total')
ax1.set_ylabel('Lucro Total ($)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Criando o segundo eixo para o Lucro Médio
ax2 = ax1.twinx()
ax2.plot(segmentos, lucro_medio, color='darkgreen', marker='o', linewidth=2, label='Lucro Médio')
ax2.set_ylabel('Lucro Médio por Pedido ($)', color='darkgreen')
ax2.tick_params(axis='y', labelcolor='darkgreen')

# Título e layout
plt.title('Lucro Total e Lucro Médio por Segmento')
fig.tight_layout()
plt.show()
```

**Análise do Lucro por Segmento**

O gráfico apresenta o Lucro Total (barras azuis) e o Lucro Médio por Pedido (linha verde) para cada segmento de clientes:

Consumer é o segmento que gera o maior lucro total, ultrapassando $730 mil, o que representa cerca de 47% do lucro total da empresa. Isso se deve principalmente ao grande volume de pedidos (26.513).

O segmento Corporate ocupa a segunda posição em lucro total, com aproximadamente $432 mil, e possui lucro médio por pedido um pouco superior ao segmento Consumer.

Apesar de ser o segmento com o menor lucro total, Home Office apresenta o maior lucro médio por pedido ($29,65), indicando que suas vendas individuais tendem a ser mais lucrativas.

**Interpretação:**

O segmento Consumer é o mais relevante em termos de volume e geração de lucro.

Já o segmento Home Office, embora menor em volume, mostra potencial para ações mais focadas, como campanhas de upselling ou fidelização.

Estratégias diferentes podem ser adotadas para cada segmento: escalar o Consumer, otimizar o Corporate, e explorar valor agregado no Home Office.

```python
pedidos_por_segmento = df['Segmento'].value_counts().reset_index()
pedidos_por_segmento.columns = ['Segmento', 'Qtd_Pedidos']
```

```python
# Plotar gráfico de barras horizontais
plt.figure(figsize=(8, 5))
sns.barplot(data=pedidos_por_segmento, x='Qtd_Pedidos', y='Segmento', palette='Blues_d')
plt.title('Número de Pedidos por Segmento')
plt.xlabel('Quantidade de Pedidos')
plt.ylabel('Segmento')
plt.show()
```

O gráfico apresenta a quantidade total de pedidos realizados por cada segmento de clientes: Consumer, Corporate e Home Office.

Os dados revelam que o segmento Consumer é o que possui o maior volume de pedidos, ultrapassando 26 mil pedidos no período analisado. Em seguida, temos o segmento Corporate, com cerca de 15 mil pedidos, e por último o segmento Home Office, com aproximadamente 9 mil pedidos.

O segmento Consumer representa a maior fatia da demanda, sendo o principal responsável pelo volume de vendas da empresa.

O segmento Home Office, apesar de ter o menor número de pedidos, apresentou anteriormente o maior ticket médio, o que pode indicar vendas de maior valor unitário.

```python
# Agrupando e calculando o ticket médio por segmento
ticket_segmento = df.groupby('Segmento').agg({
    'Vendas': ['sum', 'count']
}).reset_index()

# Renomeando colunas
ticket_segmento.columns = ['Segmento', 'Vendas_Totais', 'Qtd_Pedidos']
ticket_segmento['Ticket_Medio'] = ticket_segmento['Vendas_Totais'] / ticket_segmento['Qtd_Pedidos']

# Exibindo resultado ordenado por ticket médio
ticket_segmento.sort_values(by='Ticket_Medio', ascending=False)
```

```python
sns.barplot(data=ticket_segmento, x='Ticket_Medio', y='Segmento', palette='Greens_d')
plt.title('Ticket Médio por Segmento')
plt.xlabel('Ticket Médio ($)')
plt.ylabel('Segmento')
plt.show()
```

O gráfico apresenta o ticket médio – valor médio por pedido – para os três segmentos de clientes: Consumer, Corporate e Home Office.

Os resultados indicam que os valores de ticket médio são bastante similares entre os segmentos, todos girando em torno de $245 por pedido. O segmento Home Office apresentou o maior ticket médio, seguido de perto pelos segmentos Corporate e Consumer, praticamente empatados.

Embora o segmento Home Office tenha um ticket médio ligeiramente superior, ele representa uma fatia menor do volume total de pedidos, o que pode indicar oportunidade para ações focadas em valor agregado ou fidelização. Já os segmentos Consumer e Corporate, por terem ticket médio semelhante e geralmente maior volume de vendas, podem ser priorizados em estratégias de escala e aumento da frequência de compra.

```python
# Gráfico de dispersão
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Vendas', y='Lucro', hue='Segmento')
plt.title('Vendas vs Lucro por Segmento')
plt.xlabel('Vendas ($)')
plt.ylabel('Lucro ($)')
plt.show()
```

O gráfico mostra a relação entre o valor de vendas e o lucro obtido por pedido, segmentado por tipo de cliente: Consumer, Corporate e Home Office.
Em geral, existe uma tendência positiva: conforme as vendas aumentam, o lucro também tende a crescer.
No entanto, há muitos pontos onde vendas elevadas geraram prejuízo (lucro negativo), especialmente em pedidos com valores mais altos.
Os três segmentos apresentam essa variação, indicando que nem todo pedido com alto valor gera bom retorno financeiro.
Apesar da relação positiva entre vendas e lucro, é importante monitorar margens de lucro, pois pedidos grandes não garantem lucratividade. Estratégias de precificação e controle de custos podem ajudar a evitar prejuízos em vendas maiores.



```python
# Agrupamento por tempo de entrega (dias)
tempo_lucro = df.groupby('Tempo_Entrega_Dias')[['Lucro', 'Vendas']].mean().reset_index()

# Gráfico de linhas
plt.figure(figsize=(10, 5))
sns.lineplot(data=tempo_lucro, x='Tempo_Entrega_Dias', y='Lucro', label='Lucro Médio')
sns.lineplot(data=tempo_lucro, x='Tempo_Entrega_Dias', y='Vendas', label='Vendas Médias')
plt.title('Lucro e Vendas Médias por Tempo de Entrega')
plt.xlabel('Tempo de Entrega (dias)')
plt.ylabel('Valores Médios')
plt.legend()
plt.tight_layout()
plt.show()
```

O gráfico mostra a distribuição do tempo de entrega (em dias) para os segmentos: Consumer, Home Office e Corporate.

Todos os segmentos apresentam uma mediana de entrega de aproximadamente 4 dias.
A variação é semelhante entre os segmentos, com tempos de entrega variando de 0 a 7 dias.
Isso indica que o tempo de entrega é padronizado, sem grandes diferenças entre os perfis de clientes.
Correlação entre Tempo de Entrega e Lucro
A matriz de correlação mostra que:
Correlação entre tempo de entrega e lucro: 0.0017 (muito próximo de zero).
Interpretação:
Não há relação significativa entre o tempo de entrega e o lucro. Ou seja, entregar mais rápido ou mais devagar não afeta diretamente o lucro obtido em cada venda.
Insight:
Como o tempo de entrega não impacta o lucro, ações de melhoria logística podem ser focadas na satisfação do cliente e não necessariamente em ganhos financeiros diretos.
